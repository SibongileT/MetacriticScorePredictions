## Explore Runtime

```python
oscars_df['runtime (mins)'].value_counts()
```

```python
import datetime
def runtime_to_minutes(runtimestring):
    if 'h' in runtimestring:
        runtime = runtimestring.split('h')
        minutes = int(runtime[0])*60 
        if 'min' in runtime[1]:
            mins = runtime[1].split('min')
            minutes += int(mins[0])
    else:
        runtime_minutes = runtimestring.split('min')
        minutes = int(runtime_minutes[0])
        
    return minutes
   
```

```python
oscars_df['runtime (mins)'] = oscars_df['runtime (mins)'].apply(runtime_to_minutes)
```

```python
oscars_df['runtime (mins)'].value_counts()
```

```python
bool1 = oscars_df['runtime (mins)'] < 70
oscars_df[bool1]
oscars_df = oscars_df.drop(oscars_df[bool1].index)
```

## Create a season column based off of the date time column

```python
def get_season(column):
    month =  pd.DatetimeIndex(oscars_df[column]).month
    seasons_list = []
    for m in month:
        if m == 12 or m <= 3 :
            seasons_list.append('Winter')
        elif m <= 5 :
            seasons_list.append('Spring')
        elif m<= 8:
            seasons_list.append('Summer')
        else:
            seasons_list.append('Fall')
    return seasons_list
```

```python
oscars_df['season'] = get_season('release_date')
```

```python
oscars_df.season.value_counts()
```
