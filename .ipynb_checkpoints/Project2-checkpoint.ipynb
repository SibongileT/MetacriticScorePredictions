# Scrape IMDB

* Scraping list with every movie ever nominated for an oscar bewteen 1980 and 2020

```python
ua = UserAgent()
user_agent = {'User-agent': ua.random}
pages = np.arange(1980,2021)
title_ids=[]
for page in pages:
    url = 'https://www.imdb.com/event/ev0000003/'+str(page)
    response = requests.get(url,headers=user_agent).text
    data = json.loads( re.findall(r'IMDbReactWidgets\.NomineesWidget\.push.*?(\{.*\})', response)[0] )

    #print(json.dumps(data, indent=4)) # <-- comment this out to print all data

    for award in data['nomineesWidgetModel']['eventEditionSummary']['awards']:
        if award['awardName'] != 'Oscar':
            continue
        for title in award['categories']:
            noms = title['nominations']
            for nom in noms:
                if len(nom['secondaryNominees'])!=0:
                    id_ = nom['secondaryNominees'][0]['const']
                else:
                    id_ = nom['primaryNominees'][0]['const']
                title_ids.append(id_)
    print('-' * 80)
    sleep(randint(2,10))
len(title_ids)
```

## Scrape IMDB Title Page 
* Scrape the page of the movie to get all the necessary information from imdb 
* Director,Genre,Cast,Writer,Metacritic Score,Runtime,Release Date,Budget,Distribution Company MPAA Rating

```python
url = 'https://www.imdb.com/title/tt8991268/'
response = requests.get(url)
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
title_div = soup.find('div',class_='title_wrapper')
title_div
```

```python
writers = []
for a in soup.find('h4', text=re.compile('Writer')).parent.find_all('a'):
    writers.append(a.text.strip())
writers
```

```python
directors = []
for a in soup.find('h4', text=re.compile('Director')).parent.find_all('a'):
    directors.append(a.text.strip())
directors
```

```python
Genres = []
for a in soup.find('h4', text=re.compile('Genres')).parent.find_all('a'):
    Genres.append(a.text.strip())
Genres
```

```python
Casts = []
count = 0
for a in soup.find('h4', text=re.compile('Stars')).parent.find_all('a'):
    Casts.append(a.text)
    count+=1
    if count == 3:
        break
Casts
```

```python
metacritic_score = soup.find('div',{'class':'titleReviewBar'}).span.text
metacritic_score
```

```python
release_and_rating_string = title_div.find('div',{'class':'subtext'}).text
rating = release_and_string.split('|')[0].strip()
rating
```

```python
def get_movie_value(soup, field_name):
    
    '''Grab a value from Box Office Mojo HTML
    
    Takes a string attribute of a movie on the page and returns the string in
    the next sibling object (the value for that attribute) or None if nothing is found.
    '''
    
    obj = soup.find(text=re.compile(field_name))
    
    if not obj: 
        return None
    
    # this works for most of the values
    next_element = obj.findNext()
    
    if next_element:
        return next_element.text 
    else:
        return None
```

```python
runtime = get_movie_value(soup,'Run')
print(int(runtime.split(" ")[0]))
```

```python
title_string = get_movie_value(soup,'title')
title_string.split('(')[0].strip()
```

```python
release_and_rating_string = title_div.find('div',{'class':'subtext'}).text
release = release_and_rating_string.split('|')[3].strip()
release = release.split('(')
release = release[0]
release
```

```python
import math 
try:
    budget = soup.find('h4', text=re.compile('Budget')).parent.text
    budget = budget.split('\n')
    budget = budget[1].split('$')
    budget = budget[1].replace(',','')
    budget = int(budget)
    budget
except:
    budget = math.nan
budget
```

## Get Distribution Company From Data

```python
company_url = 'https://www.imdb.com/title/tt8991268/companycredits'
response = requests.get(company_url)
company_page = response.text
```

```python
distro_soup = BeautifulSoup(company_page, "lxml")
div_list = distro_soup.find('div',{'id':'pagecontent'})
```

```python
for div in div_list.find_all('div'):
    print(div.prettify())
```

```python
distro_list = []
usa_distribution = ''
usa_distribution_and_release = []
distribution_company = distro_soup.find('h4', {'id':'distributors'}).parent.find_all('li')
for country_distro in distribution_company:
    if 'USA' in country_distro.text:
        usa_distribution = country_distro.text
usa_distribution_and_release = usa_distribution.split('(')
usa_distribution = usa_distribution_and_release[0].replace('\n','').strip()
usa_distribution
```
