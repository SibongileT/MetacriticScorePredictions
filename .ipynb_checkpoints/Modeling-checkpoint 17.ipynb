```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso,LassoCV,LinearRegression
from sklearn.preprocessing import StandardScaler
```

## Explore the data

```python
oscars_df = pd.read_csv('oscar_movies_data_2.csv')
```

```python
oscars_df.head()
```

```python
numerics = ['int16','int32','int64','float16','float32','float64']
numerical_vars = list(oscars_df.select_dtypes(include=numerics).columns)
oscars_data = oscars_df[numerical_vars]
oscars_data.shape
```

```python
oscars_data.fillna(0,inplace=True)
```

```python
oscars_data.info()
```

## Split into test sets

```python
X = oscars_data.drop(columns=['metacritic score'])
y = oscars_data['metacritic score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=27)
```

```python
X_train.describe()
```

```python
X_test.describe()
```

## Scale Numerical Features

```python
X_train_copy = X_train.copy()
X_test_copy = X_test.copy()
```

```python
scale = StandardScaler().fit(X_train_copy[['budget']])
X_train_copy['budget'] = scale.transform(X_train_copy[['budget']])
X_test_copy['budget'] = scale.transform(X_test_copy[['budget']])
```

## Do a simple linear regression model on data

```python
lm = LinearRegression()
lm.fit(X_train_copy,y_train)
```

```python
lm.predict(X_train_copy)
```

```python
list(zip(oscars_data.drop(columns='metacritic score'),lm.coef_))
```

```python
lm.intercept_
```

```python
lm.score(X_train_copy,y_train)
```

```python
lm.score(X_test_copy,y_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lm, X, y, # estimator, features, target
                cv=5, # number of folds 
                scoring='r2') # scoring metric
```

```python
# Compute train and test errors
alphas = np.logspace(-5, 1, 60)
enet = lml.ElasticNet(l1_ratio=0.7, max_iter=10000)
train_errors = list()
test_errors = list()
for alpha in alphas:
    enet.set_params(alpha=alpha)
    enet.fit(X_train, y_train)
    train_errors.append(enet.score(X_train_copy, y_train))
    test_errors.append(enet.score(X_test_copy, y_test))

i_alpha_optim = np.argmax(test_errors)
alpha_optim = alphas[i_alpha_optim]
print("Optimal regularization parameter : %s" % alpha_optim)

# Estimate the coef_ on full data with optimal regularization parameter
enet.set_params(alpha=alpha_optim)
coef_ = enet.fit(X, y).coef_
```

```python
plt.subplot(2, 1, 1)
plt.semilogx(alphas, train_errors, label='Train')
plt.semilogx(alphas, test_errors, label='Test')
plt.vlines(alpha_optim, plt.ylim()[0], np.max(test_errors), color='k',
           linewidth=3, label='Optimum on test')
plt.legend(loc='lower left')
plt.ylim([0, 1.2])
plt.xlabel('Regularization parameter')
plt.ylabel('Performance')
```
