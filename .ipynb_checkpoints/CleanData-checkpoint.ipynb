## Read in Oscar CSV back into Dataframe

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
oscars_df = pd.read_csv('oscar_movies.csv')
```

```python
oscars_df.describe()
```

```python
oscars_df.head()
```

```python
def clean_rows(column):
    oscars_df[column] = oscars_df[column].str.strip('[]').astype(str)
    return oscars_df
```

```python
clean_rows('writer(s)')
clean_rows('director(s)')
clean_rows('casts')
clean_rows('genre(s)')
```

```python
oscars_df.head()
```

Split the writers columns and explore how many writers are in each new column

```python
oscars_df[['writers1','writers2','writers3']] = oscars_df['writer(s)'].apply(lambda x: pd.Series(x.split(',')))
```

```python
#Writers 2 should have less
oscars_df['writers2'].count()
```

```python
#Writers 3 should have less
oscars_df['writers3'].count()
```

### The third writers column is not telling us who wrote for the movie so that column will be dropped


```python
oscars_df.drop(['writer(s)','writers3'], axis=1,inplace=True)
```

```python
def runtime_to_minutes(runtimestring):
    runtime = runtimestring.split('h')
    if len(runtime)>2:
        minutes = int(runtime[0])*60 + int(runtime[2])
    else:
         minutes = int(runtime[0])*60
    return minutes
```

```python
oscars_df['runtime (mins)'] = oscars_df['runtime (mins)'].apply(runtime_to_minutes)
```
